### Dependencies and general utils

```python
import os
import numpy as np
import pdal
import json
import laspy
from tqdm import tqdm
```

### code

```python
src_folder = r"D:\GitHubProjects\Terranum_repo\TreeSegmentation\data\lausanne\tiles_vegetation"
class_name = "classification"
veg_val = [3]
extension = 'las'
```

```python
list_files = [x for x in os.listdir(src_folder) if x.endswith(extension)]
os.makedirs(os.path.join(src_folder, 'vegetation'), exist_ok=True)
for _, src_file in tqdm(enumerate(list_files), total=len(list_files)):
    src_source = os.path.join(src_folder, src_file)
    src_dest = os.path.join(src_folder, 'vegetation', src_file.split(f'.{extension}')[0] + "_veg.laz")
    laz = laspy.read(src_source)
    mask = getattr(laz, class_name) == veg_val
    laz_veg = laz[mask]
    laz_veg.write(src_dest)
```
